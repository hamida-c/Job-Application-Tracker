```python
# Copy dataset to avoid modifying original
df_model = df.copy()

# 1. Convert dates to datetime format
df_model['applied_date'] = pd.to_datetime(df_model['applied_date'],format="%d-%m-%Y", errors='coerce')
df_model['created_at'] = pd.to_datetime(df_model['created_at'], errors='coerce')

# Extract date-based features
df_model['applied_month'] = df_model['applied_date'].dt.month
df_model['applied_day'] = df_model['applied_date'].dt.day
df_model['applied_weekday'] = df_model['applied_date'].dt.weekday


```

```python
# Keep only interview and rejected statuses
df_filtered = df_model[df_model['status'].str.strip().str.lower().isin(['interview', 'rejected'])]

# Convert to binary target
df_filtered['interview_received'] = df_filtered['status'].apply(lambda x: 1 if x.strip().lower() == 'interview' else 0)

# 3. Drop unused columns (notes for now)
df_filtered = df_filtered.drop(columns=['notes', 'applied_date', 'created_at', 'status'])
```

```python
# 4. Encode categorical variables using one-hot encoding
df_model_encoded = pd.get_dummies(df_filtered, columns=['company_name', 'job_role', 'website'], drop_first=True)
```

```python
# Convert all boolean columns to int (0/1)
df_model_encoded = df_model_encoded.astype(int)
```

```python
# After encoding
df_model_encoded.columns = df_model_encoded.columns.str.strip()
```

# Modelling!!!!!!!!!



```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score
import joblib

# 1. Load your preprocessed dataset (must have interview_received as target)
df = pd.read_csv("model_ready_job_applications.csv")

# Separate features & target
X = df.drop(columns=['interview_received'])
y = df['interview_received']

# 2. Split into train/test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# 3. Train Random Forest
model_1 = RandomForestClassifier(class_weight='balanced', random_state=42)
model_1.fit(X_train, y_train)

# 4. Evaluate
y_pred = model_1.predict(X_test)
y_proba = model_1.predict_proba(X_test)[:, 1]

print(classification_report(y_test, y_pred,zero_division=0))
print("AUC Score:", roc_auc_score(y_test, y_proba))

```

```python
joblib.dump(X.columns.tolist(), "feature_columns.pkl")
print("âœ… Feature saved as feature_columns.pkl")
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, classification_report
import pandas as pd

 

# ----- Models to compare -----
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
}

# ----- Train & Evaluate -----
results = []
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else None

    results.append({
        "Model": name,
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred, zero_division=0),
        "Recall": recall_score(y_test, y_pred, zero_division=0),
        "F1 Score": f1_score(y_test, y_pred, zero_division=0),
        "AUC": roc_auc_score(y_test, y_proba) if y_proba is not None else "N/A"
    })

# ----- Show Results -----
df_results = pd.DataFrame(results)
print(df_results.sort_values(by="F1 Score", ascending=False))

```
